## Now enters Topology

```python
# transform a picture in topolofical feature
import giotto as o
from giotto.homology import VietorisRipsPersistence, CubicalPersistence
from giotto.images import DilationFiltration
from giotto.diagrams import PairwiseDistance
```

```python
# all images together
filtration = DilationFiltration()
images = np.concatenate([np.expand_dims(image, axis=0) for image in simulation.extract_slices(
    [0, np.pi/4, np.pi/2, 3*np.pi/4, np.pi, 5*np.pi/4, 3*np.pi/2, 7*np.pi/4])[0]], axis=0)
filtration.fit(images) # we take the first time step
filtrated_images = filtration.transform(images)
```

```python
cub_persistence = CubicalPersistence()
cub_persistence.fit(filtrated_images)
diagrams = cub_persistence.transform(filtrated_images)
```

```python
def plot_diagram(X, idx=0):
    """X should be a dictionary containing the homology dimesions as key, i.e. 0,1,2"""
    colors_dict = {0: 'r', 1: 'b', 2: 'y'}
    plt.figure()
    for i, key in enumerate(X[idx, :, 2]):
        key = int(key)
        plt.plot(X[idx, i,0], X[idx, i,1], 'o', color=colors_dict[key])
        
    plt.plot([0, 1.5], [0, 1.5], color='k')
    plt.show()
```

```python
plot_diagram(diagrams, 2)
```

```python
def flatten_list(list_of_list):
    flat_list = []
    for list_ in list_of_list:
        flat_list += list_
    return flat_list
```
